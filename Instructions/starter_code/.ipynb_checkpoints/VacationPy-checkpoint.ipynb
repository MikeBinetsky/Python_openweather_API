```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
filePath = os.path.join("..", "output_data", "cities.csv")
filePath
citiesDF = pd.read_csv(filePath)
citiesDF
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
locations = citiesDF[["Lat", "Lng"]]
humidity = citiesDF["Humidity"].astype(float)
```

```python
humidityFigure = gmaps.figure()
humidityHeatmap = gmaps.heatmap_layer(locations, weights=humidity)

humidityFigure.add_layer(humidityHeatmap)
humidityFigure
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
## I used the example peramaters from the instructions of temp between 70-80, wind speed less than 10, and cloudiness of 0
## Since the temperature type is in Kelvin I found the kelvin temperatures of 70 F and 80 F and plugged those in instead.
tempCitiesDF = citiesDF[(citiesDF.Temp < 299) & (citiesDF.Temp > 294)]
tempWindCitiesDF = tempCitiesDF[tempCitiesDF["Wind Speed"] < 10]
cleanCitiesDF = tempWindCitiesDF[tempWindCitiesDF.Cloudiness == 0]
cleanCitiesDF
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
## Not sure why they want us to call this a dataframe, couldn't it make more sense to have this as a list? I'm using a list.
hotel_df = cleanCitiesDF.copy()
## Using a similar method that I used in WeatherPy I'm going to store the hotels into this list, then add them to the hotels_df
HOTELS = []
placeCode = []
```

```python
## The type of data that the LAT and LNG columns are need to be converted to a string to be called for my API URL
hotel_df["Lat"] = hotel_df["Lat"].astype(str)
hotel_df["Lng"] = hotel_df["Lng"].astype(str)
```

```python
## Here is the API call parameters
api_key = g_key
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"
params = "&radius=5000&keyword=hotel"

## This loop goes through the API calls and store the relevant data in a list
for index, row in hotel_df.iterrows():
    response = requests.get(base_url + "&location=" + row["Lat"] + "," + row["Lng"] + params + "&key=" + api_key)
    responseJSON = response.json()
    try:
        HOTELS.append(responseJSON["results"][0]["name"])
    except:
        HOTELS.append("No Hotel within 5000m")
    try:
        placeCode.append(responseJSON["results"][0]["plus_code"]["compound_code"])
    except:
        placeCode.append("N/A")
```

```python
## Adding the new columns to the dataframe
hotel_df["Hotel Name"] = HOTELS
hotel_df["Country"] = placeCode
```

```python
## Correcting the country name to include only the country's name
hotel_df["Country"] = hotel_df["Country"].str.split(',').str[1]
```

```python
hotel_df["Lat"] = hotel_df["Lat"].astype(float)
hotel_df["Lng"] = hotel_df["Lng"].astype(float)
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
humidityFigure.add_layer(markers)
# Display figure
humidityFigure
```
